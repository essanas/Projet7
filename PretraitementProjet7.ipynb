### <a name="C3">3 - Prétraitement du texte avec NLP</a>

```python
# Importations pour le traitement du langage naturel
import nltk
import string
from nltk.stem import WordNetLemmatizer

# Téléchargement des ressources nécessaires pour nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('words')
nltk.download('wordnet')
```

```python
# Liste des mots vides (stop words) en anglais
stopwords = nltk.corpus.stopwords.words('english')
# Ensemble de mots en anglais
words = set(nltk.corpus.words.words())
# Initialisation de l'objet lemmatizer pour la lemmatisation des mots
lemmatizer = WordNetLemmatizer()
```

```python
def Preprocess_listofSentence(listofSentence):
    # Initialisation de la liste pour stocker les phrases prétraitées
    preprocess_list = []

    # Parcours de chaque phrase dans la liste d'entrée
    for sentence in listofSentence:
        # Retire la ponctuation et convertit en minuscules
        sentence_w_punct = "".join([i.lower() for i in sentence if i not in string.punctuation])

        # Supprime les chiffres de la phrase
        sentence_w_num = ''.join(i for i in sentence_w_punct if not i.isdigit())

        # Tokenization de la phrase
        tokenize_sentence = nltk.tokenize.word_tokenize(sentence_w_num)

        # Supprime les stopwords de la phrase
        words_w_stopwords = [i for i in tokenize_sentence if i not in stopwords]

        # Lemmatization des mots
        words_lemmatize = (lemmatizer.lemmatize(w) for w in words_w_stopwords)

        # Crée une phrase propre en assemblant les mots lemmatisés
        sentence_clean = ' '.join(w for w in words_lemmatize if w.lower() in words or not w.isalpha())

        # Ajoute la phrase propre à la liste des résultats
        preprocess_list.append(sentence_clean)

    # Retourne la liste des phrases prétraitées
    return preprocess_list
```

```python
# Appliquer la fonction de prétraitement aux données textuelles de la colonne 'tweet' du DataFrame 
preprocess_list = Preprocess_listofSentence(data_unique['tweet'])
```

```python
data_unique['Tweet_corrige'] = Preprocess_listofSentence(data_unique['tweet'])
```
